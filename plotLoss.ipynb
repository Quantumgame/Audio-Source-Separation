```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time 
```

```python
with open('./lossRecord/lossfileNewNet.txt') as f:
#with open('./lossRecord/lossfiletwo.txt') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
plotcon = np.around(np.array(content).astype(float),2)
```

```python
sns.set_style("darkgrid")
plt.plot(plotcon)
plt.xticks()
plt.yticks()
plt.show()
```

```python
while True:
    sns.set_style("darkgrid")
    plt.plot(plotcon[:])
    plt.xticks()
    plt.yticks()
    plt.show()
    time.sleep(1) 
    plt.close()
```
