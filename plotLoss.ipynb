```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time 
```

```python
with open('./lossRecord/newmodellossfile.txt') as f:
#with open('./lossRecord/lossfiletwo.txt') as f:
    content = f.readlines()
content = [x.strip() for x in content] 
plotcon = np.around(np.array(content).astype(float),2)
```

```python
while True:
    with open('./lossRecord/newmodellossfile.txt') as f:
    #with open('./lossRecord/lossfiletwo.txt') as f:
        content = f.readlines()
    content = [x.strip() for x in content] 
    plotcon = np.around(np.array(content).astype(float),2)
    sns.set_style("darkgrid")
    plt.plot(plotcon)
    plt.xticks()
    plt.yticks()
    plt.show()
    time.sleep(100) 
```
